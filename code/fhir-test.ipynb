```python
from langchain.llms import HuggingFacePipeline
import transformers
import torch
import accelerate

import pandas as pd
from langchain.prompts import FewShotPromptTemplate
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

from transformers import AutoTokenizer #AutoTokenizer. A tokenizer is responsible for preprocessing text into an array of numbers as inputs to a model.

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Llama 2 Model - 7B fine-tuned model, optimized for dialogue use cases and converted for the Hugging Face Transformers format
# Llama 2 was trained between January 2023 and July 2023.

model="meta-llama/Llama-2-7b-chat-hf" # try with meta-llama/Llama-2-7b which is not chat-hf

tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
# this takes time (3-5 minutes)

pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
# this takes time (3-5 minutes) - addressing issue of max_length that happened when I used the previous cell

pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    # Instead of max_length, use max_new_tokens to control the generated output length
    max_new_tokens=200, # Adjust this value as needed
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0.1})
```

```python
def few_shot_symptom(text, examples):
    example_prompts = "\n\n".join([f"text: {ex['text']}\nnegation: {ex['negation']}" for ex in examples])
    prompt = f"Based on the examples below, determine if the symptom or medication in the given clinical document is negated. Respond with only 'yes' or 'no'. Do not include any explanations or extra text.\n\nExamples:\n{example_prompts}\n\ntext: {text}\nnegated:"

    response = llm(prompt).strip().lower()

    # Ensure only "yes" or "no" is returned
    if response not in {"yes", "no"}:
        response = "no"  # Default fallback to avoid errors

    return response

examples = [
    {
        "text": "The patient was given acetaminophen and complained of headaches but no fatigue.",
        "ner": "acetaminophen",
        "type": "medication",
        "negation": "no"
    },
    {
        "text": "The patient was given acetaminophen and complained of headaches but no fatigue.",
        "ner": "fatigue",
        "type": "symptom",
        "negation": "yes"
    },
    {
        "text": "She did not experience itching.",
        "ner": "itching",
        "type": "symptom",
        "negation": "yes"
    }
]

df['negation'] = df['text'].apply(lambda x: few_shot_symptom(x, examples))
df.head()
```
