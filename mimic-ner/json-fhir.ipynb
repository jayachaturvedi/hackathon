```python
import json

# Define your mapping: {old_key: new_key}
column_mapping = {
    "old_heading_1": "new_heading_1",
    "old_heading_2": "new_heading_2",
    "old_heading_3": "new_heading_3"
}

# Load the JSON file
def load_json(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        return json.load(file)

# Rename keys in the JSON data
def rename_keys(data, mapping):
    if isinstance(data, dict):
        return {mapping.get(k, k): rename_keys(v, mapping) for k, v in data.items()}
    elif isinstance(data, list):
        return [rename_keys(item, mapping) for item in data]
    else:
        return data

# Save the modified JSON file
def save_json(data, output_path):
    with open(output_path, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

# Example usage
input_file = "input.json"      # Replace with your input file path
output_file = "output.json"    # Replace with your desired output file path

json_data = load_json(input_file)
updated_json = rename_keys(json_data, column_mapping)
save_json(updated_json, output_file)

print(f"Updated JSON saved to {output_file}")
```
